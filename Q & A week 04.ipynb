```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("Telco.xlsx")
data.head()
```

# 1.Compare churn count with respect to gender.

```python
sns.countplot(data['gender'],hue=data['Churn'],data=data)
plt.title('Churn count vs Gender', fontweight='bold', fontsize=10)
plt.grid()
```

# 2.Find out how many female senior citizens there in the dataset


```python
plt.figure(figsize=(8,4))
f=data[data['gender']=='Female']
s=f[f['SeniorCitizen']==1.0]
a=s[['gender','SeniorCitizen']]
g=len(a['SeniorCitizen'])
sns.countplot(x='gender',data=a)
plt.grid()
```

# 3.Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(10,10))
plt.scatter(data['tenure'],data['TotalCharges'])
plt.title('Tenure vs Total Charges', fontsize=15)
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
```

# 4. Find out which contract is preferred by the senior citizen

```python
senior=data[data['SeniorCitizen']==1.0]
senior
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=senior['Contract'])
plt.title('Contract preference of Senior Citizen', fontweight='bold', fontsize=15)
plt.grid()
plt.show()
```

# 5. Comment on your finds on Payment Method

```python
plt.figure(figsize=(10,10))
sns.countplot(x=data['PaymentMethod'])
plt.title('Prefered payment method', fontweight='bold', fontsize=18)
plt.xticks()
plt.grid()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=data['PaymentMethod'], hue=data['Contract'])
plt.title('Prefered payment method; contractwise', fontweight='bold', fontsize=20)
plt.xticks()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=data['PaymentMethod'], hue=data['PaperlessBilling'])
plt.title('Prefered payment method Vs billing', fontweight='bold', fontsize=20)
plt.xticks()
plt.grid()
plt.show()
```
